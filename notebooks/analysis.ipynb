```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load merged dataset
df = pd.read_csv("../data/processed/pga_stats_merged.csv")

df.head()
```

```python
# Remove % and convert to float
for col in ["DrivingAccuracy", "GIR"]:
    df[col] = df[col].str.replace("%", "").astype(float)

# Convert others to float too, just in case
df["DrivingDistance"] = df["DrivingDistance"].astype(float)
df["ScoringAverage"] = df["ScoringAverage"].astype(float)

df.describe()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="DrivingDistance", y="ScoringAverage")
plt.title("Driving Distance vs Scoring Average")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Between Golf Stats")
plt.show()
```
